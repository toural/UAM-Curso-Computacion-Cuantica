```python
%matplotlib inline

from qiskit import *
from qiskit.visualization import *
from qiskit.tools.visualization import plot_histogram
from qiskit_textbook.tools import array_to_latex
import numpy as np

qr = QuantumRegister(3)
cr = ClassicalRegister(3)

qc = QuantumCircuit(qr,cr)

qc.x(qr[0])
# qc.x(qr[1])
qc.x(qr[2])

qc.h(qr[2])
qc.cu1(np.pi/4,qr[2],qr[0])
qc.cu1(np.pi/2,qr[2],qr[1])

qc.h(qr[1])
qc.cu1(np.pi/2,qr[1],qr[0])

qc.h(qr[0])
qc.swap(qr[0],qr[2])

qc.draw()

# state vector

stvBackend = Aer.get_backend('statevector_simulator')
job = execute (qc, stvBackend)

state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)
```

```python
qc.draw()
```

```python
plot_bloch_multivector(state)
```

```python
unitBackend = Aer.get_backend('unitary_simulator')
job = execute (qc, unitBackend)

state = job.result().get_unitary()
array_to_latex(state, pretext="\\text{Unitary} = ", precision=1)
```
