```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram

q = QuantumRegister(2)
c = ClassicalRegister(2)
circ = QuantumCircuit(q, c)

circ.h(q[0])
circ.cx(q[0], q[1])
circ.measure(q[0], c[0])


emulator = Aer.get_backend('qasm_simulator')
job = execute( circ, emulator, shots=10 ) 

# The result is a histogram in the form of a dictionary.

histogram = job.result().get_counts()
print ('results: ', histogram)

# plot histogram

legend = ['Execution results']
plot_histogram(histogram,legend=legend)
```

```python
job2 = execute( circ, emulator, shots=100 )
histogram2 = job2.result().get_counts()
legend = ['First execution', 'Second execution']
plot_histogram([histogram, histogram2],legend=legend)
```

```python
job3 = execute( circ, emulator, shots=8192 )
histogram3 = job3.result().get_counts()
legend = ['First', 'Second', 'Third']
plot_histogram([histogram, histogram2, histogram3], legend=legend)
```

```python
print ("# RESULTS MEMORY   #############################################################")

# list ordered results, not just aggregated results

job = execute( circ, emulator, shots=100, memory=True )
samples = job.result().get_memory()
print ('samples: ', samples)
```
