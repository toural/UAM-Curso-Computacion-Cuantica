```python
from qiskit import *
from qiskit.converters import circuit_to_dag
from qiskit.tools.visualization import dag_drawer

qr = QuantumRegister(1,'qr')
cr = ClassicalRegister(1, 'cr')
qc1 = QuantumCircuit(qr, cr, name='ghz')


qc1.draw()
```

```python
qc1_dag = converters.circuit_to_dag(qc1)
dag_drawer(qc1_dag,scale=0.8)
```

```python
qc1.measure(qr,cr)
qc1.draw()

```

```python
qc1_dag = converters.circuit_to_dag(qc1)
dag_drawer(qc1_dag,scale=0.8)
```

```python
qr = QuantumRegister(2,'qr')
cr = ClassicalRegister(2, 'cr')
qc1 = QuantumCircuit(qr, cr, name='ghz')

qc1.h(qr[0])

qc1.measure(qr,cr)

qc1.draw()
```

```python
qc1_dag = converters.circuit_to_dag(qc1)
dag_drawer(qc1_dag,scale=0.8)
```

```python
qr = QuantumRegister(2,'qr')
cr = ClassicalRegister(2, 'cr')
qc1 = QuantumCircuit(qr, cr, name='ghz')

qc1.h(qr[0])
qc1.cx(qr[0],qr[1])

qc1.measure(qr,cr)

qc1.draw()
```

```python
qc1_dag = converters.circuit_to_dag(qc1)
dag_drawer(qc1_dag,scale=0.8)
```

```python
qr = QuantumRegister(3,'qr')
cr = ClassicalRegister(3, 'cr')
ghz = QuantumCircuit(qr, cr, name='ghz')
ghz.h(qr[0])
ghz.x(qr[0])
ghz.ccx(qr[0],qr[1],qr[2])

ghz.measure(qr,cr)

ghz.draw()
```

```python
ghz_dag = converters.circuit_to_dag(ghz)

print(ghz.width(), ghz_dag.width())

dag_drawer(ghz_dag,scale=0.8)
```
