```python
%matplotlib inline
from qiskit import *
from qiskit import IBMQ
from qiskit import Aer
import numpy as np

# IBMQ.update_account()

IBMQ.load_account()

print ("IBMQ providers: ")
for provider in IBMQ.providers():
    print(provider)
    
provider = IBMQ.get_provider(hub='ibm-q')
print (provider)

print ("IBMQ backends: ")
for backend in provider.backends(open_pulse=True):
    print(backend)

print(backend)
```

```python
qr = QuantumRegister (1)
cr = ClassicalRegister (1)
qc = QuantumCircuit (qr,cr)

qc.h(qr[0])


qc.measure(qr,cr)

qc.draw(output='mpl')


armonk = backend

qc_armonk = transpile(qc,armonk)

qc_armonk.draw(output='mpl')

scheduled_circ = schedule(qc_armonk,armonk)

scheduled_circ.draw()
```
