```python
%matplotlib inline

from qiskit import *
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.visualization import plot_state_paulivec, plot_state_hinton
from qiskit.visualization import plot_state_qsphere
from qiskit_textbook.tools import array_to_latex
import numpy as np
```

```python
qr = QuantumRegister(1)

circ = QuantumCircuit(qr)

circ.draw()
```

```python
for backend in Aer.backends():
    print(backend.name())
```

```python
stvBackend = Aer.get_backend('statevector_simulator')

job = execute (circ, stvBackend)
state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)
```

```python
circ.x(qr)

job = execute (circ, stvBackend)
state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)
```

```python
circ.h(qr)

job = execute (circ, stvBackend)
state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)
```

```python
qr = QuantumRegister(3)

qc = QuantumCircuit(qr)

qc.h(qr[0])
qc.x(qr[1])
# qc.u3(np.pi/3,np.pi/7,np.pi/7,qr[2]) check

qc.u3(np.pi/3,np.pi,np.pi/7,qr[2])

job = execute (qc, stvBackend)
state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)

qc.draw()
```

```python
plot_bloch_multivector(state,title="Blochsphere")
```
