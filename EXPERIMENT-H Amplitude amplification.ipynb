## Subroutine: Amplitude Amplification

```python
from qiskit import *
from qiskit_textbook.tools import array_to_latex


qr = QuantumRegister(5)
cr = ClassicalRegister(5)
qc = QuantumCircuit(qr,cr)

qc.h(qr[0:4])
qc.barrier()

# state to be marked e.g. 2
qc.x(qr[2:4])
qc.x(qr[0])
qc.barrier()

qc.ccx(qr[0],qr[1],qr[4])
qc.h(qr[3])
qc.ccx(qr[2],qr[4],qr[3])
qc.h(qr[3])
qc.ccx(qr[0],qr[1],qr[4])

# state to be marked e.g. 2
qc.barrier()
qc.x(qr[2:4])
qc.x(qr[0])

qc.draw()
```

```python
stvBackend = Aer.get_backend('statevector_simulator')

job = execute (qc, stvBackend)

state = job.result().get_statevector()
array_to_latex(state, pretext="\\text{Statevector} = ", precision=1)
```

```python
total_prob = 0
for i,amp in enumerate(state):
    if abs(amp) > 0.000001:
        prob = abs(amp) * abs(amp)
        total_prob += prob
       # print('|{}> {} probability = {}%'.format(i, amp, round(prob * 100, 5)))
        print('|{}> probability = {}%'.format(i, round(prob * 100, 5)))
print('Total probability: {}%'.format(int(round(total_prob * 100))))
```

```python
from qiskit.tools.visualization import plot_histogram

qc.measure(qr[0:4],cr[0:4])

emulator = Aer.get_backend('qasm_simulator')
job = execute(qc, emulator, shots=10000 ) 

histogram = job.result().get_counts()
print (histogram)
plot_histogram(histogram)
```
